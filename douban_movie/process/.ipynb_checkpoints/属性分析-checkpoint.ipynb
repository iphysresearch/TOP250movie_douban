```python
def get_info(arr, df, dict_topic):
#     topic_arr = ['逻辑','节奏','手法','细节','个性','情节','剧情','呈现方式','个性','创意','想法','剧本','原著','掌控能力','能力','功利','水平','风格']
    total = 0
    for topic in topic_arr:
        flag = df.content.str.contains(topic)
#         print('------------{}-------------------'.format(topic))
#         print('总数', len(df[flag]))
#         print('满意度数量', len(df[flag].loc[df['Sentiment']=='positive']))
#         print(len(df[flag].loc[df['Sentiment']=='positive']), len(df[flag]))
        dict_topic[topic] = [len(df[flag]), round(len(df[flag].loc[df['Sentiment']=='positive'])/len(df[flag]),2)]
        total += len(df[flag])
    print('total', total)
    return total, dict_topic
    
```

## 导演创作

```python
topic_arr = ['逻辑','节奏','团队','调度','手法','细节','情节','剧情','呈现方式','个性','创意','想法','剧本','原著','掌控能力','能力','功力','水平','风格','场面']
dict_topic = {}
total, dict_topic = get_info(topic_arr, df, dict_topic)
print(total)
for topic in topic_arr:
    print('-------------{}---------------------------'.format(topic))
    print('关注数量', dict_topic[topic][0])
    print('满意数量', dict_topic[topic][1])
    print('{}关注度'.format(topic), round(dict_topic[topic][0]/total, 2))
    print('{}满意度'.format(topic), dict_topic[topic][1])
    print('----------------------------------------')
```
